##피벗 테이블

##넓은 테이블 형식에서 긴 테이블 형식으로 변환
<img width="700" alt="image" src="https://pandas.pydata.org/docs/_images/07_melt.svg">

melt() 메소드
* id_vars: 기준
* value_vars: 행으로 보낼 열
* var_name: 행으로 보내진 열의 열 이름
* value_name: 값의 열 이름

reset_index()를 통해 새로운 인덱스를 추가 하여 넓은 테이블 형식 생성

```python
class3_pivoted = titanic.pivot(columns="sex", values="fare").reset_index()
class3_pivoted
```

```python
class3_melt = class3_pivoted.melt(id_vars="index")
class3_melt
```

```python
class3_melt = class3_pivoted.melt(id_vars="index", value_vars=["male"], value_name="Fare", var_name="Male")
class3_melt
```

#**시계열 데이터를 쉽게 처리하는 방법**

##시간 시리즈를 다른 주파수로 재샘플링
